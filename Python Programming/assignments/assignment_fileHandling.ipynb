### 2. Write a Python program to read first n lines of a file.

```python
n = int(input("Enter number of line to be printed"))
cnt = 0
with open('filehandelling.txt') as f:
    for lines in f:
        print(lines)
        cnt +=1
        if cnt == n:
            break
```

```python
n = int(input("Enter the number of lines to be printed"))
with open('filehandelling.txt') as f:
    for lines in (f.readlines()[:n]):
        print(lines)
```

```python
n = int(input("Enter line number to be printed"))
with open('filehandelling.txt') as f:
    line = f.readlines()
    print(line[n])
```

### 4. Write a Python program to read last n lines of a file.

```python
n = int(input("write last n lines to be printed"))
with open('filehandelling.txt') as f:
    for lines in (f.readlines()[-n:]):
        print(lines)
```

### 5. Write a Python program to read a file line by line and store it into a list.

```python
l1 =[]
with open('filehandelling.txt') as f:
    for lines in f.readlines():
        #print(type(lines))
        l1.append(lines)
print(l1)
len(l1)
```
