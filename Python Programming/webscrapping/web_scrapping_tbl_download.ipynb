# Get all table tags

```python
all_table = soup.find_all('table')
```

```python
interested_tbl = []
for tbl in all_table:
    if "Rank" in tbl.text:
#         print("#########")
#         print(tbl)
        all_tr = tbl.find_all('tr') # table row tag
        for row in all_tr:
#             print("$$$$$$$$$")
#             print(row.text)
            row_contents = row.text.split("\n")
            interested_tbl.append(row_contents)
#             print(row_contents)

import pandas as pd
df = pd.DataFrame(interested_tbl)
df.to_csv("tbl.csv")
```

```python
# Get all URLs within the tables of the page
links =[]
for table in all_table:
#     print("\n\n\n")
    all_a = table.find_all('a', href=True)
    for ref in all_a:
        links.append(ref['href'])
# print("Count = ",len(links))
# print(links)
```
